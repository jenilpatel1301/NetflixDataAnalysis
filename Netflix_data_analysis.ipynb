## Q2. In which year highest number of the TV shows & Movies were released ?

```python
df['Date_R'] = pd.to_datetime(df['Release_Date'])
```

```python
df.head()
```

```python
df['Date_R'].dt.year.value_counts().plot(kind='bar')
```

In year 2019 highest number of movies and Tv shows were released
